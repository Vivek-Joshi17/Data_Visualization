```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
exam_perf = pd.read_csv('student_exam_data.csv')
exam_perf
```

# Plotting betweem result and study hours

```python
plt.scatter(exam_perf['Study Hours'],exam_perf['Pass/Fail'],cmap='winter',alpha=0.7)
plt.xlabel('Study Hours')
plt.ylabel('Result')
plt.colorbar()
```

**Plotting between result and Study hours is not conclusive enough**

```python
plt.scatter(exam_perf['Study Hours'],exam_perf['Previous Exam Score'],c=exam_perf['Pass/Fail'],cmap='cool',alpha=1)
plt.xlabel('Study Hours')
plt.ylabel('Previous Exam Score')
plt.colorbar()
plt.show()
```

**In conclusion upper right corner are the student who passed the exam**

```python
row_index = 2
selected_row = exam_perf.iloc[:]

temp = selected_row.values
```

```python
temp.shape
```

```python
min_study=15
for i in range(500):
  if(temp[i][0]< min_study and temp[i][2]==1):
    min_study = temp[i][0]
```

```python
min_study
```

```python
min_score=100
for i in range(500):
  if(temp[i][1]< min_score and temp[i][2]==1):
    min_score = temp[i][1]
```

```python
min_score
```

```python
plt.figure(figsize=(12,12))
plt.scatter(exam_perf['Study Hours'],exam_perf['Previous Exam Score'],c=exam_perf['Pass/Fail'],cmap='cool',alpha=1)
plt.xlabel('Study Hours')
plt.ylabel('Previous Exam Score')
plt.axhline(min_score,color='red')
plt.axvline(min_study,color='red')
plt.colorbar()
plt.show()
```

**The upper right region is of  student who passed the exam**
- In conclusion the student should have scored approx 60 marks in previous exam and should study for atleast 5 hours to pass the exam
