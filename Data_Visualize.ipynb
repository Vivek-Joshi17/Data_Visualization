```python
#Dataset
batsman = pd.read_csv("/content/sharma-kohli.csv")
batsman
```

```python
#2D plot

plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
#Plotting two column at a time
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
#Title and label

plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('R Sharma vs V Kohli')
plt.xlabel('season')
plt.ylabel('score')
```

```python
#colors and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='red')
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('R Sharma vs V Kohli')
plt.xlabel('season')
plt.ylabel('score')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='red')
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dashed')
plt.title('R Sharma vs V Kohli')
plt.xlabel('season')
plt.ylabel('score')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='red')
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dotted')
plt.title('R Sharma vs V Kohli')
plt.xlabel('season')
plt.ylabel('score')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='red')
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dashdot')
plt.title('R Sharma vs V Kohli')
plt.xlabel('season')
plt.ylabel('score')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='red')
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dashdot',linewidth=3)
plt.title('R Sharma vs V Kohli')
plt.xlabel('season')
plt.ylabel('score')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='red',marker='D')
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dashdot',linewidth=3,marker='+',markersize=15
         )
plt.title('R Sharma vs V Kohli')
plt.xlabel('season')
plt.ylabel('score')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='red',marker='D',label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dashdot',linewidth=3,marker='+',markersize=15,label='Rohit'
         )
plt.title('R Sharma vs V Kohli')
plt.xlabel('season')
plt.ylabel('score')

plt.legend()
```

```python
#Outlier is present
price =[48000,54000,57000,49000,47000,45000,4500000]
year=[2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
```

```python
#Trimming the graph
price =[48000,54000,57000,49000,47000,45000,4500000]
year=[2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
#trimming the graph
plt.ylim(0,75000)
```

```python
price =[48000,54000,57000,49000,47000,45000,4500000]
year=[2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
#trimming the graph
plt.ylim(0,74000)
plt.xlim(2016,2019)
plt.grid()
```

```python
price =[48000,54000,57000,49000,47000,45000,4500000]
year=[2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
#trimming the graph
plt.ylim(0,74000)
plt.xlim(2016,2019)
plt.grid()
plt.show()
```
